```python
import numpy as np

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

```python
#reading data
data = pd.read_csv("ds_salaries.csv")
data
```

```python
#feature_names
data.keys()
```

```python
#we are checking data's value
data.isnull().sum()
```

```python
data
```

```python
#kelimelerin sayıya dönüşmüş halini almamızı sağlar

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
#yukarıdaki kodun kullanım şekli
data["job_title"] = encoder.fit_transform(data["job_title"])
data["experience_level"] = encoder.fit_transform(data["experience_level"])
data["employee_residence"] = encoder.fit_transform(data["employee_residence"])
data["employment_type"] = encoder.fit_transform(data["employment_type"])
data["company_location"] = encoder.fit_transform(data["company_location"])
data["salary_currency"] = encoder.fit_transform(data["salary_currency"])
data = data.replace({"company_size":{"L":1,"M":3,"S":3}})



```

```python
data
```

```python
#splitting data    
x = data.drop(columns=(["salary_in_usd","Unnamed: 0","salary"]),axis=1)
y = data["salary_in_usd"]
```

```python
#model creating step 1
x_train,x_test,y_train_,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#model creating step 2
from xgboost import XGBRegressor
model = XGBRegressor()
model.fit(x_train,y_train_)
```

```python
#2 prediction model
prediction = model.predict(x_test)
prediction2 = model.predict(x_train)
```

```python
#r values
from sklearn import metrics
r2 = metrics.r2_score(y_test,prediction)
r22 = metrics.r2_score(y_train_,prediction2)
print(r2)
print(r22)

```
